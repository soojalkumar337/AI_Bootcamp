```python
# Task 2

score = int(input("Enter Your Score: "))
if score >= 80:
    grade = 'A1'
elif score >= 70:
    grade = 'A'
elif score >= 60:
    grade = 'B'
elif score >= 50:
    grade = 'C'
elif score >= 40:
    grade = 'D'
else: 
    grade = 'Fail'
print(grade)    
```

```python
# Task 6
text = "Python is simple yet powerful. Python is easy to learn."
# Split text and count frequency
words = text.split()
word_freq = {}
# Fill the logic here to count words
for word in words:
    if word in word_freq:
        word_freq[word] += 1
    else:
        word_freq[word] = 1
print(word_freq)
```
