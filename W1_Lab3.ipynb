```python
# Task 1        loads the data

import pandas as pd 
df = pd.read_csv("hourly_weather_10_days.csv")
print(df.head(5))   # display few rows that you want 
print(f"The no total no of rows and columns is{df.shape}")      # display total rows and columns
```

```python
# Task 2           Basic Exploration
print(df.info())   # display column names and data types

print(f"\n Basic Statistics: \n{df.describe()}")   # disply basic statistics

print(df.isna().sum())     # count missing values in each column
```

```python
# Task 3   Handle Missing Values

#df['temperature_C'] = df['temperature_C'].fillna(df['temperature_C'].mean())
#df['humidity_%'] = df['humidity_%'].fillna(df['humidity_%'].mean())
#df['wind_speed_kmph'] = df['wind_speed_kmph'].fillna(df['wind_speed_kmph'].mean())
#df['pressure_hPa'] = df['pressure_hPa'].fillna(df['pressure_hPa'].mean())
#df['visibility_km'] = df['visibility_km'].fillna(df['visibility_km'].mean())


# By using loop handle missing values in columns
for col in df.columns:
    if df[col].dtype != 'object':
        df[col] = df[col].fillna(df[col].mean())
print(df.isna().sum())        
```

```python
# Task 4 Data Analysis
df['timestamp'] = pd.to_datetime(df['timestamp'])
daily_avg_temp = df.groupby(df['timestamp'].dt.date)['temperature_C'].mean()
print(daily_avg_temp)

print(df[['temperature_C']].agg(['min', 'max', 'mean']))
print(df[['humidity_%']].agg(['min', 'max', 'mean']))
print(df[['wind_speed_kmph']].agg(['min', 'max', 'mean']))
print(df[['pressure_hPa']].agg(['min', 'max', 'mean']))
print(df[['visibility_km']].agg(['min', 'max', 'mean']))

df['hour'] = df['timestamp'].dt.hour
avg_humidity_by_hour = df.groupby('hour')['humidity_%'].mean()
print(avg_humidity_by_hour)
```

```python
# Task 5 Numpy Matrix Exercises 
import numpy as np

temp = df['temperature_C'].to_numpy()
wind = df['wind_speed_kmph'].to_numpy()
print(f"Temp{temp} \n Wind{wind}")
```

```python
# a) Reshape into matrix form 
temp_matrix = temp.reshape((10, 24))
print(f"Temp Matrix after reshaping:\n{temp_matrix}")
# axis=0 → Columns
# axis=1 → Rows

daily_min = temp_matrix.min(axis=1)
print(f"Daily Min Temp: {daily_min}")
daily_max = temp_matrix.max(axis=1)
print(f"Daily Max Temp: {daily_max}")
daily_mean = temp_matrix.mean(axis=1)
print(f"Daily Mean Temp: {daily_mean}")
```

```python
# b) Normalize the temp matrix
# Subtract the mean and divide by std deviation
# Do it manually using NumPy functions
import numpy as np

def normalize(matrix):
    sub_matrix= matrix - np.mean(matrix)
    normalized_temp_matrix = sub_matrix / np.std(matrix) 
    return normalized_temp_matrix
print(normalize(temp_matrix))

```

```python
# c) Apply custom mask/filter
# Create a mask for wind speed > 15 kmph
# Use it to extract high-wind readings
mask = wind > 15
high_wind = wind[mask]
print(high_wind)
```

```python
# Task 6 Fianl Task 
#  Write and test your function
import numpy as np

def daily_summary(matrix):
    summaries = []

    for day in matrix:
        summaries.append({
            "min": np.min(day),
            "max": np.max(day),
            "mean": np.mean(day)
        })

    return summaries


# Test the function
summaries = daily_summary(temp_matrix)

for i, day in enumerate(summaries, start=1):
    print(f"Day {i}: {day}")
```
